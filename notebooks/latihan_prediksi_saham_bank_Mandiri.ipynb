### CATATAN PENGOLAHAN SAHAM ML BANK MANDIRI

Ini merupakan catatan saya untuk menemukan model untuk memprediksi saham yang terjadi di Bank Mandiri. Penelitian saya di Bank BRI, Mega, dan BCA lalu, hanya menggunakan metode regresi linear. Kali ini, saya akan mencoba menggunakan hal yang berbeda dengan metode yang namanya adalah *Facebook Prophet*

---

#### Apa itu metode *Facebook Prophet*?

*Facebook Prophet* adalah model forecasting (peramalan) yang dikembangkan oleh tim Data Science Meta (Facebook).
Berbeda dengan Regresi Linear yang saya pakai sebelumnya (yang hanya menarik garis lurus kaku), Prophet bekerja lebih manusiawi. Ia melihat data waktu (time series) sebagai gabungan dari tiga elemen:

- Trend                   : Arah utama pergerakan (sedang naik, turun, atau datar).
- Seasonality (Musiman)   : Pola yang berulang (mingguan, bulanan, atau tahunan).
- Holidays (Efek Liburan) : Kejadian khusus yang berpotensi mengguncang data (misal: Lebaran, Tahun Baru, atau event ekonomi).

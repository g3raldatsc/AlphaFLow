## CATATAN PENGOLAHAN SAHAM ML BANK MANDIRI

Ini merupakan catatan saya untuk menemukan model untuk memprediksi saham yang terjadi di Bank Mandiri. Penelitian saya di Bank BRI, Mega, dan BCA lalu, hanya menggunakan metode regresi linear. Kali ini, saya akan mencoba menggunakan hal yang berbeda dengan metode lainnya.
